```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft, fftfreq, ifft
```

```python
n = 1000
Lx = 100
ω = 2.0 * np.pi/Lx

```

```python
#Creating the Signals#

x = np.linspace(0,Lx,n)
y1 = 1.0*np.cos(5.0*ω*x)
y2 = 2.0*np.sin(10.0*ω*x)
y3 = 0.5*np.cos(20.0*ω*x)

#full signal#

y = y1+y2+y3
```

```python
ν = fftfreq(n)
mask = ν > 0
```

```python
fft_vals = fft(y)
```

```python
fft_theo = 2.0*np.abs(fft_vals/n)
```

```python
plt.figure(1)
plt.title('original signal')
plt.plot(x,y,color='xkcd:salmon', label='original')
plt.legend()
```

```python
plt.figure(2)
plt.title('raw fft values')
plt.plot(ν,fft_vals)
```

```python
plt.figure(2)
plt.title('raw fft values')
plt.plot(ν[mask],fft_theo[mask])
```
