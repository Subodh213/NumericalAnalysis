```python
import matplotlib.pyplot as plt
from numpy import*

def fn(t):
    return(exp(-t))
def exact(t):
    return(-exp(-t))
def Trapezoid(A,B,n):
    h = (B-A)/(n-1)
    sum_trap = (fn(A)+fn(B))/2
    for i in range(1,n-1):
        sum_trap += fn(A+(i*h))
    sum_trap *=h
    return (sum_trap)
A = 0
B = 1
N = 200
exact_sum = exact(B)-exact(A)
Et =[]
It=[]
print("N", "\t", "Trapezoid Area", '\t', "Relative Error", "\n")
for j in range(2,N):
    err_t = abs((Trapezoid(A,B,j)-exact_sum)/exact_sum)
    Et.append(err_t)
    It.append(j)
    print(j, "\t", "%.4E"%Trapezoid(A,B,j), "\t","%.4E"%err_t)
print("\n Print Exact Area is =", "%.4E"%exact_sum)
    
#plt.plot(It, Et)
plt.loglog(It,Et)
plt.title("A plot of relative errors versus N")
plt.xlabel("N")
plt.ylabel("Relative Error")
plt.grid()
plt.show()
    
```
